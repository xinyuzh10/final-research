```python
train_data = total_consumption.iloc[:15,2:]
train_spending =  total_consumption.iloc[:15,1]

predict_data_covid = total_consumption.iloc[15:42,2:]
predict_data_post_covid = total_consumption.iloc[42:,2:]
```

```python
print(train_data)
print(train_spending)
print(predict_data_covid)
print(predict_data_post_covid) 
```

```python
train_spending.hist(bins=15)
```

```python
model = LinearRegression()
```

```python
model.fit(train_data, train_spending)
```

```python
r_sq = model.score(train_data, train_spending)
print(f"coefficient of determination: {r_sq}")
print(f"intercept: {model.intercept_}")
print(f"slope: {model.coef_}")
```

```python
y_pred = model.predict(predict_data_covid)
```

```python
print(f"predicted response:\n{y_pred}")
```

```python
y_pred = model.predict(predict_data_post_covid)
```
